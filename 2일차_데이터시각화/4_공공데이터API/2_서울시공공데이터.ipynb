```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df = pd.read_csv("서울시 공영주차장 안내 정보.csv", encoding="cp949")
df
```

```python
df.info()
```

```python
df['주소']
```

```python
ku_dict = {}
```

```python
for temp in df['주소']:
    ku = temp.split()[0]
    print(ku)
```

```python
for temp in df['주소']:
    ku = temp.split()[0]
    if ku in ku_dict:
        ku_dict[ku] = ku_dict[ku] + 1
        # ku_dict[ku] += 1
    else:
        ku_dict[ku] = 1
print(ku_dict)
```

```python
ku_df = pd.DataFrame.from_dict([ku_dict])
ku_df
```

```python
ku_df = ku_df.T
ku_df
```

```python
ku_df.plot(kind='bar')
plt.show()
```
